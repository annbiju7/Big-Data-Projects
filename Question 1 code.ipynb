```
import re
from operator import add
from pyspark import SparkContext

# File location and type
file_location = "/FileStore/tables/input_hw1_3_.txt"
file_type = "txt"

# loading input file into the RDD
words_rdd = sc.textFile("/FileStore/tables/input_hw1_3_.txt")

#preprocessing
rdd1 = words_rdd.flatMap(lambda x:x.lower().split(" ")).filter(lambda x:x!='')

#map rdd
map_rdd = rdd1.map(lambda word:(word, 1))

# RBK
rdd2 = map_rdd.reduceByKey(lambda x, y: x + y)
rdd2.collect()

# sorting
rdd2.sortBy(lambda b:b[1]).collect()
display(rdd2.take(10))
```
